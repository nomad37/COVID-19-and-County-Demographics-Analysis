```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
country = 'United States'
age = df[(df['State'] == country) & (df['Sex'] == 'All')].sort_values('COVID-19 Deaths')
age
```

```python
plt.barh(y = age['Age group'], width = age['COVID-19 Deaths']);
plt.title('Older Age Groups Are at higher risk from Dieing to COVID-19')
```

```python
plt.figure(figsize = (12, 6))
plt.barh(y = age['Age group'], width = age['Total Deaths']);
plt.title('Older Age Groups Are at higher risk from Dieing to COVID-19')
```

```python
plt.figure(figsize = (12, 6))
plt.barh(y = age['Age group'], width = age['Pneumonia Deaths']);
plt.title('Older Age Groups Are at higher risk from Dieing to COVID-19')
```

```python
plt.figure(figsize = (12, 6))
plt.barh(y = age['Age group'], width = age['Influenza Deaths']);
plt.title('Older Age Groups Are at higher risk from Dieing to COVID-19')
```

```python
country = 'United States'
sex = df[(df['State'] == country)].sort_values('COVID-19 Deaths')
sex = sex.groupby('Sex').sum()
sex
```

```python
plt.bar(x = sex.index, height = sex['COVID-19 Deaths']);
plt.title('Males are at a slightly higher risk')
```
